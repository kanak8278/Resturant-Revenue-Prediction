```python
cont_vars= ['Id',
 'P2',
 'P3',
 'P4',
 'P13',
 'P26',
 'P27',
 'P28',
 'P29',
 'Open Week',
 'Open Day',
 'Open Dayofyear',
 'Open Elapsed']
```

```python
df = train_df[cat_vars + cont_vars + [dep_var, 'Open Date'] ].copy()
```

```python
type_list = cont_cat_split(train_df)
type_list[0]
# type_list[1]
```

```python
train_df[['revenue', 'P1', 'P2', 'P3', 'P4', 'P28', 'P35']].corr()
```

```python
valid_idx= range(len(train_df)-100, len(train_df))
```

```python
procs = [FillMissing, Categorify, Normalize]
```

```python
dep_var = 'revenue'
categorical = type_list[1]
cont = type_list[0]
```

```python
data = (TabularList.from_df(path ='./data/', df= df, cat_names= cat_vars, cont_names= cont_vars, procs= procs)
       .split_by_idx(valid_idx)
       .label_from_df(cols=dep_var, label_cls= FloatList, log=False)
       .add_test(TabularList.from_df(test_df, path='./data/', cat_names= categorical, cont_names= cont, procs= procs))
       .databunch(bs=32)
       )
```

```python
max_y = (np.max(train_df['revenue'])*1.2)
y_range = torch.tensor([0, max_y], device=defaults.device)
```

```python
learn = tabular_learner(data, layers=[1000,500], ps=[0.001,0.01], emb_drop=0.05, 
                        y_range=y_range, metrics=rmse)
```

```python
learn.model
```

```python
learn.lr_find(start_lr= 1e-2, end_lr= 100)
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, 1e-2)
```

```python
preds, y =learn.get_preds(DatasetType.Test)
```
